```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("End-to-End Business Intelligence Dashboard Started Successfully")
```

```python
orders = pd.read_csv("../data/orders.csv")
customers = pd.read_csv("../data/customers.csv")
products = pd.read_csv("../data/products.csv")

print("Orders Data")
print(orders.head())

print("\nCustomers Data")
print(customers.head())

print("\nProducts Data")
print(products.head())
```

```python
df = pd.merge(orders, customers, on="Customer_ID")
df = pd.merge(df, products, on="Product_ID")

print(df.head())
```

```python
print(df.info())
```

```python
total_sales = df["Sales"].sum()

print("Total Sales =", total_sales)
```

```python
total_profit = df["Profit"].sum()

print("Total Profit =", total_profit)
```

```python
total_customers = df["Customer_ID"].nunique()

print("Total Customers =", total_customers)
```

```python
sales_product = df.groupby("Product_Name")["Sales"].sum()

sales_product.plot(kind="bar", figsize=(8,5))

plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")

plt.show()
```

```python
profit_product = df.groupby("Product_Name")["Profit"].sum()

profit_product.plot(kind="bar", figsize=(8,5))

plt.title("Profit by Product")
plt.xlabel("Product Name")
plt.ylabel("Profit")

plt.show()
```

```python
region_sales = df.groupby("City")["Sales"].sum()

region_sales.plot(kind="bar", figsize=(8,5))

plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Sales")

plt.show()
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))

plt.title("Sales by Category")

plt.show()
```
